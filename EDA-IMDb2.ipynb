```python
listGenre = []
for i in IMDb_movies["genre"]:
    array = i.split(",")
    for j in array:
        listGenre.append(j)

liste =list(dict.fromkeys(listGenre))
ListFinal = []
for x in liste:
     l = x.replace(' ', '')
     ListFinal.append(l)
kFinal = list(dict.fromkeys(ListFinal))
kFinal
```

```python
ListeDistribution = []

for t  in kFinal:
    i = 0
    for m in IMDb_movies["genre"]:
        o = t in m
        
        if o == True:
           i = i +1
    ListeDistribution.append({t:i})
        
ListeDistribution
```

```python
listkey = []
listvalue= []
for item in ListeDistribution:
    for key , value in item.items():
        listkey.append(key)
        listvalue.append(value)

import pandas as pd

data = {'genre':  listkey,
        'counting': listvalue,
        
        }

df = pd.DataFrame(data)

print(df)
```

```python
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
import seaborn as sns 

plt.figure(figsize=(12,10))
splot=sns.barplot(x="genre",y="counting",data=df,ci=None)
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.0f'),
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center',
                   xytext = (0, 10), 
                   textcoords = 'offset points')
    
plt.xticks(rotation=90, size=12)
plt.xlabel("Genre", fontsize=15)
plt.ylabel("Counting", fontsize=15)
plt.title("Nombre de films produits par genre", size=18)
plt.plot()
```

```python
ListeDistributions= []

for t in kFinal:
    rating=0
    
    for indx , m in enumerate(IMDb_movies["genre"]):
        o = t in m
        
        if o == True:
           rating = rating + IMDb_movies["avg_vote"][indx]
    ListeDistributions.append({t:rating})
        
ListeDistributions
```

```python
listkeys = []

listvalues= []

listkey = []
listvalue= []
for item in ListeDistribution:
    for key , value in item.items():
        listkey.append(key)
        listvalue.append(value)
        
        
for item in ListeDistributions:
    for key , value in item.items():
        listkeys.append(key)
        listvalues.append(value)
import pandas as pd

datas = {'genre':  listkey,
        'counting': listvalue,
        'avgrating' : listvalues
        
        }

dfs = pd.DataFrame (datas)

print(dfs)

dfs['avg'] = dfs['avgrating']/dfs['counting']
dfs
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

plt.figure(figsize=(12,10))
splot=sns.barplot(x="genre",y="avg",data=dfs,ci=None)
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.1f'),
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 10), 
                   textcoords = 'offset points')
plt.xticks(rotation=90)
plt.xlabel("Genre", fontsize=15)
plt.ylabel("Average", fontsize=15)
plt.title("RÃ©partition des notes de films pour chaque genre", size=18)
plt.plot()
```

```python
listYear = []
for i in IMDb_movies["year"]:
   listYear.append(i)
listYear
```

```python
list_obs_annee=[]
for element in IMDb_movies["year"]:
    if element=="TV Movie 2019":
         list_obs_annee.append(2019)
    else:
         list_obs_annee.append(int(element))
   
listYear=list(set(list_obs_annee))
listYear=sorted(listYear)

def compt_annee(liste,annee):
    i=0
    for element in liste:
        if element==annee :
            i=i+1
    return i
 
ListeDistribution2 = []    
for element in listYear:
    compt=compt_annee(list_obs_annee,element)
    ListeDistribution2.append({element:compt})
ListeDistribution2
```

```python
listkey2 = []
listvalue2 = []
for item in ListeDistribution2:
    for key , value in item.items():
        listkey2.append(key)
        listvalue2.append(value)
import pandas as pd

data = {'annee':  listkey2,
        'nb films produits': listvalue2,
        
        }

df = pd.DataFrame (data)

print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

plt.figure(figsize=(30,12))
splot=sns.barplot(x="annee",y="nb films produits",data=df,ci=None)

plt.xticks(rotation=90, size=13)
plt.yticks(size=15)
plt.xlabel("AnnÃ©e", fontsize=25)
plt.ylabel("Nombre de films produits", fontsize=25)
plt.title("Ã‰volution du nombre de films produits au cours des annÃ©es", size=30)
plt.show()
```
